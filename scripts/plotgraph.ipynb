```python
import igraph
from igraph import plot
import numpy as np
```

```python
g = igraph.Graph()

with open('../data/cora.metis') as file:
	for i, line in enumerate(file):
		if line == '\n':
			continue
		if i == 0:
			verts, edges = line.split(' ')
			verts = int(verts)
			g.add_vertices(list(range(verts)))
		
		else:
			neighbors = [int(node) for node in line.split(' ')]

			for n in neighbors:
				g.add_edge(i - 1, n - 1)
				


```

```python
clustering = []
with open('../data/cora.cluster') as file:
	for line in file:
		vert, cluster = line.split(' ')
		vert, cluster = int(vert) - 1, int(cluster)
		clustering.append(cluster)
		
np.unique(clustering)
```
