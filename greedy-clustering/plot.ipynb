```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.dpi'] = 300
```

```python
mods_lux = []
with open('luxembourg-modularities') as file:
	for line in file:
		mods_lux.append(float(line))
```

```python
mods_cora = []
with open('cora-modularities') as file:
	for line in file:
		mods_cora.append(float(line))
```

```python
mods_cora_normalized = []
with open('cora_normalized-modularities') as file:
	for line in file:
		mods_cora_normalized.append(float(line))
```

```python
mods_333SP = []
with open('333SP-modularities') as file:
	for line in file:
		mods_333SP.append(float(line))
```

```python
mods_rgg_17 = []
with open('rgg_17-modularities') as file:
	for line in file:
		mods_rgg_17.append(float(line))
```

```python
plt.plot(mods_lux)
plt.xlabel('thresholds')
plt.ylabel('modularity')

print(np.max(mods_lux))
```

```python
plt.plot(mods_cora, label='not normalized')
plt.plot(mods_cora_normalized, label='normalized')
plt.title('cora greedy clustering')
plt.xlabel('thresholds')
plt.ylabel('modularity')

print(np.max(mods_cora))
```

```python
plt.plot(mods_cora, label='cora')
print(np.max(mods_cora))
plt.plot(mods_333SP, label='333SP')
print(np.max(mods_333SP))
plt.plot(mods_rgg_17, label='rgg_17')
print(np.max(mods_rgg_17))


plt.xlabel('thresholds')
plt.ylabel('modularity')
plt.legend(loc='upper right')
```
